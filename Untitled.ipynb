# Import libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# URL

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250',headers=headers).text
```

# Beautiful Soup object

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
list_movie=soup.find_all('li',class_="ipc-metadata-list-summary-item")
```

```python
name=[]
year=[]
rating=[]
length=[]
reviews=[]
for i in list_movie:
    name.append(i.find('h3').text.split('. ')[1])
    year.append(i.find_all('span',class_='cli-title-metadata-item')[0].text)
    rating.append(i.find('span',class_='ipc-rating-star').text.split('\xa0')[0])
    reviews.append(i.find('span',class_='ipc-rating-star').text.split('\xa0')[1].lstrip('(').rstrip(')'))
    length.append(i.find_all('span',class_='cli-title-metadata-item')[1].text)
    
```

```python
length
```

```python
df = pd.DataFrame({'name':name,'year':year,'rating':rating,'length':length,'reviews':reviews})
```

```python
df.to_csv('filename.csv',index=False)
```
